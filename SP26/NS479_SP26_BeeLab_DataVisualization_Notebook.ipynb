## Flipping the data format

This data is in **wide format**, where each row would represent a different bee and all data for that bee would be contained in the columns of a single row. <br>
For working with the plotting package (`seaborn`), our data needs to be in **long format**. This means we need to transform the data so that an individual bee is actually repeated across multiple rows of the dataset.<br>
A single bee is repeated across the rows. The data for each bee will appear in multiple rows, with different columns containing info about the trial types:
* `CS_Type`: `CS_Plus` and `CS_Minus`
* `Measure`: `Choice` and `Expected`
<br>

You don't have to understand all of the details in the following code, it is meant for you to run and put the data into the format that you need for visualization.

````python
#@title Just run this cell - hidden code

## Convert to long format
df_long = bees.melt(id_vars="Bee",var_name="Category", value_name="Outcome")
## Extract CS type (Plus/Minus) and measure (Chose/Expected)
df_long["CS_Type"] = df_long["Category"].apply(lambda x: "CS_Plus" if "Plus" in x else "CS_Minus")
df_long["Measure"] = df_long["Category"].apply(lambda x: "Choice" if "Choice" in x else "Expected")
## Reorganize columns
df_long = df_long[["CS_Type", "Measure", "Outcome"]]
df_long
````

# Plot behavior for each individual bee side-by-side 👥

````python
#@title Pivot the table and create a new column to categorize behavior

# Pivot to wide format so each trial is one row with CS_Plus and CS_Minus outcomes
trial_df = training_long.pivot_table(
    index = ["Bee Name", "Trial", "Training_Phase"],
    columns = "CS_Type",
    values = "Outcome"
).reset_index()

# Fill missing choices with 0 (if any)
trial_df = trial_df.fillna(0)

# Create a categorical column for the choice pattern
def trial_pattern(row):
    if row['CS_Plus'] == 1 and row['CS_Minus'] == 0:
        return "Plus Only"
    elif row['CS_Plus'] == 0 and row['CS_Minus'] == 1:
        return "Minus Only"
    elif row['CS_Plus'] == 1 and row['CS_Minus'] == 1:
        return "Both"
    else:
        return "Neither"

trial_df['Pattern'] = trial_df.apply(trial_pattern, axis=1)
````

````python
#@title Plot each individual bee - plotting function

def plot_all_bees(
    trial_df,
    n_cols = 5,
    save_dir = None,
    fig_name = "All_Bees_Choice_Patterns.png",
    plot_title = "Bee Choice Patterns Over Trials"
):
    """
    Plot all bees' choice patterns in a grid layout.

    Parameters:
    - trial_df: DataFrame containing trial-level data
    - n_cols: number of subplot columns
    - save_dir: optional directory to save figure
    - fig_name: filename for saved figure
    - plot_title: overall title for the figure
    """

    # ----- Pattern mapping (bottom → top order) -----
    pattern_order = ["Neither", "Minus Only", "Both", "Plus Only"]
    pattern_map = {p: i for i, p in enumerate(pattern_order)}

    df = trial_df.copy()
    df["Pattern_num"] = df["Pattern"].map(pattern_map)

    # ----- Grid layout -----
    all_bees = df["Bee Name"].unique()
    n_rows = math.ceil(len(all_bees) / n_cols)

    fig, axes = plt.subplots(
        n_rows, n_cols,
        figsize=(4 * n_cols, 3 * n_rows),
        sharey=True
    )
    axes = axes.flatten()

    # ----- Color palette -----
    phases = df["Training_Phase"].unique()
    palette = sns.color_palette("colorblind", n_colors=len(phases))

    # Manual legend handles
    handles = [
        mpatches.Patch(color = palette[i], label=phase)
        for i, phase in enumerate(phases)
    ]

    # ----- Plot each bee -----
    for i, bee in enumerate(all_bees):
        bee_df = df[df["Bee Name"] == bee]

        sns.scatterplot(
            data = bee_df,
            x = "Trial",
            y = "Pattern_num",
            hue = "Training_Phase",
            palette = palette,
            s = 80,
            ax = axes[i],
            legend = False
        )

        axes[i].set_title(bee)
        axes[i].set_xlabel("Trial")
        axes[i].set_ylabel("")
        axes[i].set_yticks(range(len(pattern_order)))
        axes[i].set_yticklabels(pattern_order)
        axes[i].set_ylim(-0.3, len(pattern_order) - 1 + 0.3)

    # Remove empty subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    # ----- Figure-level legend -----
    fig.legend(
        handles = handles,
        title = "Training Phase",
        bbox_to_anchor = (1.05, 0.5),
        loc = "center left",
        borderaxespad = 0
    )

    fig.suptitle(plot_title, y = 1.02)

    plt.tight_layout()
    plt.subplots_adjust(right = 0.85)

    # ----- Save if requested -----
    if save_dir is not None:
        os.makedirs(save_dir, exist_ok=True)
        full_path = os.path.join(save_dir, fig_name)
        fig.savefig(full_path, dpi = 300, bbox_inches = "tight")
        print(f"Saved figure to: {full_path}")

    plt.show()
````

````python
### RUN THIS CELL ###

plot_all_bees(trial_df,save_dir = save_dir)
````

````python
#@title Option 3 - Point Plot for Training Data - plotting function

def plot_individual_bee_point(bee_name, df, jitter=0.1, plot_title=None, fig_name=None, save_dir=None):
    """
    Plot training behavior for a single bee using a point/strip style plot.

    Parameters:
    - bee_name: str, name of the bee to plot
    - df: DataFrame, the full training_long dataframe
    - jitter: float, horizontal jitter for individual points (0 = no jitter)
    - plot_title: str, optional title for the figure
    - fig_name: str, optional file name to save the figure
    - save_dir: str, optional file path to save the figure
    """

    # Filter for the selected bee
    bee_df = df[df["Bee Name"] == bee_name]

    if bee_df.empty:
        print(f"No data found for bee: {bee_name}")
        return

    # Set default title if none provided
    if plot_title is None:
        plot_title = f"Training Behavior Over Time: {bee_name}"

    # Set default filename if none provided
    if fig_name is None:
        fig_name = f"PointPlot_{bee_name}.png"

    # Determine full path for saving
    if save_dir is not None:
        os.makedirs(save_dir, exist_ok=True)  # create folder if needed
        full_path = os.path.join(save_dir, fig_name)
    else:
        full_path = fig_name

    # Create the pointplot (mean/summary)
    g = sns.catplot(
        data = bee_df,
        x = "Trial",
        y = "Outcome",
        hue = "CS_Type",
        col = "Training_Phase",
        capsize = .2,
        palette = "colorblind",
        kind = "point",
        height = 3,
        aspect = 0.75
    )

    # Overlay the individual points with jitter, if requested
    if jitter > 0:
        for ax, phase in zip(g.axes[0], bee_df['Training_Phase'].unique()):
            phase_df = bee_df[bee_df['Training_Phase'] == phase]
            sns.stripplot(
                data = phase_df,
                x = "Trial",
                y = "Outcome",
                hue = "CS_Type",
                palette = "colorblind",
                dodge = True,
                jitter = jitter,
                alpha = 0.6,
                ax = ax,
                legend=False
            )

    # Adjust aesthetics
    g.set(yticks = [0,1])
    g.set_titles(col_template = "Training Phase {col_name}")
    g.set_ylabels("Chosen")
    g.fig.suptitle(plot_title, y = 1.05)


    # Save figure
    plt.savefig(full_path, dpi = 300, bbox_inches = "tight")
    plt.show()

    print(f"Saved figure to: {full_path}")
````

````python
#@title Option 4 - Strip Plot for Training with Binary (Yes/No) Outcome - plotting function


def plot_individual_bee_strip_binary(bee_name, df, jitter=False, plot_title=None, fig_name=None,save_dir=None):
    """
    Plot a stripplot for an individual bee, showing their training behavior over trials.

    Parameters:
    - bee_name: str, name of the bee to plot
    - df: DataFrame, full dataframe with training data
    - jitter: bool, whether to add horizontal jitter to points
    - plot_title: str, optional plot title
    - fig_name: str, optional filename to save the figure
    - save_dir: str, optional file path to save the figure
    """

    # Filter for the selected bee
    bee_df = df[df["Bee Name"] == bee_name]

    if bee_df.empty:
        print(f"No data found for bee: {bee_name}")
        return

    # Default title and filename
    if plot_title is None:
        plot_title = f"Training Behavior Over Time: {bee_name}"
    if fig_name is None:
        fig_name = f"StripPlot_{bee_name}.png"

    # Determine full path for saving
    if save_dir is not None:
        os.makedirs(save_dir, exist_ok=True)  # create folder if needed
        full_path = os.path.join(save_dir, fig_name)
    else:
        full_path = fig_name

    # Draw stripplot
    g = sns.catplot(
        data = bee_df,
        x = "Trial",
        y = "Binary",
        hue = "Training_Phase",
        col = "CS_Type",
        palette = "viridis",
        kind = "strip",
        height = 3,
        aspect = 0.75,
        marker = "o",
        **{"s": 100, "jitter": jitter}
    )

    # Adjust aesthetics
    g.set_titles(col_template = "{col_name}")
    g.set_ylabels("Chosen")
    g.set(xticks=sorted(bee_df["Trial"].unique()))  # show all x-axis values
    g.fig.suptitle(plot_title, y = 1.05)


    # Save figure
    plt.savefig(full_path, dpi = 300, bbox_inches = "tight")
    plt.show()

    print(f"Saved figure to: {full_path}")
````
